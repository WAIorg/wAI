```python
import cv2
import numpy as np
import open3d as o3d
```

```python
depth_image = './images/sub-002/depth1.png'
depth_img = cv2.imread(depth_image, cv2.IMREAD_UNCHANGED)
print(depth_img.shape, depth_img.dtype)

# convert to greyscale
gray = cv2.cvtColor(depth_img, cv2.COLOR_BGR2GRAY)
```

```python
# normalize values
depth_sim = gray.astype(np.float32) / 255.0

# image size
h, w = depth_sim.shape

# estimates of camera depth info
fx = fy = 500  # focal length
cx = w / 2
cy = h / 2

# creates a point cloud from depth
i, j = np.meshgrid(np.arange(w), np.arange(h))
z = depth_sim.flatten()
x = (i.flatten() - cx) * z / fx
y = (j.flatten() - cy) * z / fy
points = np.vstack((x, y, z)).T

# create Open3D point cloud
pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(points)

# color points based on original image
colors = cv2.cvtColor(depth_img, cv2.COLOR_BGR2RGB)
pcd.colors = o3d.utility.Vector3dVector(colors.reshape(-1, 3) / 255.0)

# visualize
o3d.visualization.draw_geometries([pcd])

# save
o3d.io.write_point_cloud("pointcloud.ply", pcd)
```
